```python
df = pd.read_csv("related.csv").sort_values(["Year"])
df = df[["Paper", "Features", "Co-location", "Flow", "Convoy", "Swarm", "Distributed"]].dropna()
df = df[df["Paper"].apply(lambda x: "BerminghamL15" not in x and "bao2021mining" not in x )]
df
```

```python
gdf = df.groupby(by=["Features", "Co-location", "Flow", "Convoy", "Swarm", "Distributed"]) \
        .apply(lambda x: ', '.join(list(x["Paper"]))).reset_index()
gdf
```
